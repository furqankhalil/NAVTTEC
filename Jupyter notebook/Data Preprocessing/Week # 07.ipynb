#   Data Cleaning
##  Missing value imputation using Scikit-Learn
### for Numeric and Categorical Variables/Data

```python
import numpy as np
import pandas as pd 
from sklearn.impute import SimpleImputer
```

```python
train = pd.read_csv(r"/home/abbas/NAVTTEC/Dataset_files/DataCleaning/houseprizes/train.csv")
test = pd.read_csv(r"/home/abbas/NAVTTEC/Dataset_files/DataCleaning/houseprizes/test.csv")
print("shape of train df = ",train.shape)
print("shape of test df = ",test.shape)
```

```python
train.head()
```

```python
#In ML Algo we have test and train
#X is capital because it consider it as matrix
X_train=train.drop(columns="SalePrice") #we drop saleprice because it is our target value
y_train=train["SalePrice"] #it is a vector
print("shape of X_train df = ",X_train.shape)
print("shape of y_train df = ",y_train.shape)
```

# Numerical Missing Value Imputation

```python
#We will take all the numeric variables
num_vars=X_train.select_dtypes(include=["int64","float64"]).columns
```

```python
num_vars
```

```python
#To Check the null value
X_train[num_vars].isnull().sum()
```

```python
# we will create object of class
imputer_mean = SimpleImputer(strategy='mean') #strategy we can take mode median
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)#if we want to fill by constant value
```

```python
#Take the dataset and find mean of numerical variables
imputer_mean.fit(X_train[num_vars]) #Fit method will take mean of num_var and store it as a blue print
```

```python
#T0 check the mean value in variables
imputer_mean.statistics_
```

```python
#to impute the value in x-train dataset we will use transform()
imputer_mean.transform(X_train[num_vars]) #it will give us an arrray  which is disadvantage  of sklearn

#we give dataset and it return 2d array 
```

```python
#to convert it into data frame
X_train[num_vars] = imputer_mean.transform(X_train[num_vars])
test[num_vars] = imputer_mean.transform(test[num_vars])#when we impute train data we will also have to impute test data
```

```python
X_train[num_vars].isnull().sum()
```

```python
test[num_vars].isnull().sum()
```

# Categorical Missing Value Imputation

```python
cat_vars=X_train.select_dtypes(include=["O"]).columns
cat_vars
```

```python
X_train[cat_vars].isnull().sum()
```

```python
imputer_mode = SimpleImputer(strategy='most_frequent')
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
imputer_mode
```

```python
imputer_mode.fit(X_train[cat_vars])
```

```python
#To check the mode value in each variables/coloumn

imputer_mode.statistics_
```

```python
X_train[cat_vars] = imputer_mode.transform(X_train[cat_vars])
test[cat_vars] = imputer_mode.transform(test[cat_vars])
```

```python
X_train[cat_vars].isnull().sum()
```

```python
test[cat_vars].isnull().sum()
```
