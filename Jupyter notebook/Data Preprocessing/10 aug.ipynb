```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('E:\\Study\\Python\\csvfiles\\tips.csv')
df.head()
```

```python
df.shape
```

```python
df.keys()
```

```python
df2 = df[['total_bill', 'tip','size']]
df2
```

```python
X=df2.drop('total_bill',axis=1)
y=df2['total_bill']
print(X.shape,'\n',y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split (X,y,test_size=0.2,random_state=13)
#this will give use four data sets
print(X_train.shape,"\n",X_test.shape,"\n",y_train.shape,"\n",y_test.shape)
```

```python
sc=StandardScaler()
sc.fit(X_train)
#StandardScaler will learn S.D and Mean from data
```

```python
sc.mean_
```

```python
sc.scale_
```

```python
X_train.describe()
#it gives us the details of numerical features
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
print(X_train_sc,"\n\n",X_test_sc)
```

```python
X_train_sc=pd.DataFrame(X_train_sc,columns=['tip','size'])
X_test_sc=pd.DataFrame(X_train_sc,columns=['tip','size'])
```

```python
print(X_train_sc)
X_test_sc
```
