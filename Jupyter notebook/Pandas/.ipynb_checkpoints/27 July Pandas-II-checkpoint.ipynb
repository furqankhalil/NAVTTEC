# Data Frame

```python
import pandas as pd
```

```python
#Creating using list of list

list_of=[[1,2,3,4],[2,4,6,8],[3,4,5,7]]

df1=pd.DataFrame(list_of)
df1
```

```python
#Creating using list in dict

dc1={'ID': [23,43,41],
    'SN':[0,1,2]}
df2=pd.DataFrame(dc1)
print(df2)
```

```python
#Creating using list of dict

ls_dt=[
    {'a':1,'b':2},
    {'a':3,'b':5,'c':7}

]
df3=pd.DataFrame(ls_dt)
print(df3)
```

```python
#Creating DataFrame using Dict of Series
ds={'ID': pd.Series([23,43,41]),
    'SN':pd.Series([0,1,2])}
df4=pd.DataFrame(ds)
print(df4)
```

```python
#Reading a csv file

df9=pd.read_csv("csvfiles/Data.csv")
df9
```

```python
#Selecting no of rows in csv file

df9=pd.read_csv("csvfiles/Data.csv",nrows=5)
df9
```

```python
#Selecting no of col in csv file

df9=pd.read_csv("csvfiles/Data.csv",usecols=[0,1])
df9
```

```python
#For skipig specific rows

df9=pd.read_csv("csvfiles/Data.csv",skiprows=[2])
df9
```

```python
#For selecting index col

df9=pd.read_csv("csvfiles/Data.csv",index_col='Age')
df9
```

```python
#For Header

df9=pd.read_csv("csvfiles/Data.csv", header=None)
df9
```

```python
#Parameters of Pandas read_csv Funtions:
```

```python
#Prefix= to change header using string
df9=pd.read_csv("csvfiles/Data.csv",prefix='column',header=None)
df9
```

```python
#names= to give names to the columns using list names=['a','b','c','d']

df9=pd.read_csv("csvfiles/Data.csv",names=['a','b','c','d'])
df9
```
